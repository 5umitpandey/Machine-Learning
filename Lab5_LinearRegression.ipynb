```python
from sklearn.model_selection import train_test_split
```

```python
##Step3  Build the model
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
##Step4  Fit the model ( Train the model)
```

```python
model.fit(xtrain, ytrain)
```

```python
###Step 5 Testing the model ( prediction from test data)
ypred= model.predict(xtest)

```

```python
ypred
```

```python
ytest
```

```python
## step 6  Evaluation
```

```python
from sklearn.metrics import r2_score
R2score = r2_score(ytest, ypred)
R2score
```

```python
## we have another metrics also, before that we will plot
```

```python
#plot the line of regression (training)
plt.scatter(xtrain, ytrain, color ='blue', s =50, label='actual points')
plt.scatter(xtrain, model.predict(xtrain), color ='red', s =50, label='Predicted points')
plt.plot(xtrain, model.predict(xtrain), color ='orange', label='Predicted points', linestyle ='dotted')
plt.legend()
plt.show()
```

```python
##plot for testing data
plt.scatter(xtest, ytest, color ='blue', s =50, label='actual points')
plt.scatter(xtest, model.predict(xtest), color ='red', s =50, label='Predicted points')
plt.plot(xtest, model.predict(xtest), color ='orange', label='Predicted points', linestyle ='dotted')
plt.legend()
plt.show()
```

```python
## for which random state we have the best fit
```

```python
scores = []
for i in range(20000):
  xtrain1,xtest1,ytrain1,ytest1 = train_test_split(x,y,train_size=0.8,random_state=i)
  model1 = LinearRegression()
  model1.fit(xtrain1,ytrain1)
  ypred1 = model1.predict(xtest1)
  r2_1 = r2_score(ytest1,ypred1)
  scores.append(r2_1)
```

```python
np.argmax(scores)
```
