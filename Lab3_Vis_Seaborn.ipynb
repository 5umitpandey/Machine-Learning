```python
#Relationship plot
```

```python
sns.relplot(x='total_bill',y='tip',s=100,data =tips,hue='sex')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',s=50,data =tips,hue='day')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',s=100,data =tips,hue='day',col='day', col_wrap =4)
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',s=200,data =tips,hue='time',col='time',col_wrap=2)
plt.show()
```

```python
tips.info()
```

```python
### Joint Plot = Scatter + Histogram
sns.jointplot(x='total_bill',y='tip',data=tips,s=200)
plt.show()
```

```python
### Multiple graphs
a = sns.PairGrid(tips)
a.map_diag(sns.histplot)
a.map_offdiag(sns.scatterplot)
plt.show()
```
