```python
data.dropna(axis=0, how = "any", subset = ['umpire1','umpire2'])
```

```python
### fill the null values
```

```python
mean_runs = round(data['win_by_runs'].mean())
data['win_by_runs'] = data['win_by_runs'].fillna(mean_runs)
```

```python
data = data.fillna("NP")
```

```python
data_null = pd.DataFrame(data.isnull().sum())
import seaborn as sns
sns.heatmap(data_null,annot=True,cmap='Blues')
```
