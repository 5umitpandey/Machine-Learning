```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
```

```python
# Load the sample dataset
df = pd.read_csv("Iris.csv")
```

```python
df
```

```python
df.shape
```

```python
# Standardize the data
X = df[['SepalLength', 'SepalWidth','PetalLength', 'PetalWidth']].values
X = (X - np.mean(X, axis=0)) / np.std(X, axis=0)
X
```

```python
# Standardize the data
#P= df[['SepalLength', 'SepalWidth','PetalLength', 'PetalWidth']]
#P = (P - np.mean(P, axis=0)) / np.std(P, axis=0)
#P
```

```python
# Calculate Principal Components
pca = PCA()
Z = pca.fit_transform(X)
Z
```

```python
Y=pd.DataFrame(Z)
Y
```

```python
# Access the explained variance ratio for each Principal Component
explained_variance = pca.explained_variance_ratio_

print("Explained variance for each Principal Component:")
print(explained_variance)
```

```python
pca = PCA(n_components=3)

```

```python
# Fit and transform the data
X_reduced = pca.fit_transform(Y)
```
